```python
import pandas as pd
```

```python
stats = pd.read_csv('DemographicData.csv')
```

```python
stats
```

```python
stats.columns = ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers', 'IncomeGroup']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize']= 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
#Distribution:
visl1 = sns.distplot(stats['InternetUsers'])
```

```python
#Boxplots:
vis2 = sns.boxplot(data=stats, x='IncomeGroup', y='BirthRate')
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 10)
```

```python
##Keyword Arguments in Python (advanced tutorial)
```

```python
#Marker size
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 8, 
                  scatter_kws={'s':100})
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 8, 
                  scatter_kws={'s':200})
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 8, 
                  scatter_kws={'s':50})
```

```python
# Doesn't use marker field in scatter_kws
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 8, 
                  scatter_kws={'s': 50, 'marker':'s'})
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 8, 
                  markers=['s', 'o', '^', 'D'], scatter_kws={'s': 50})
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate',
                  fit_reg = False, hue='IncomeGroup', size = 8, 
                  markers=['s', 'o', '^', 'D'], scatter_kws={'s': 75})
```
